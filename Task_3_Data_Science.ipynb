```python
#Count Plots for all colums
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(15,7))
sns.countplot(df['Ship Mode'],ax=axs[0][0])
sns.countplot(df['Segment'],ax=axs[0][1])
sns.countplot(df['Region'],ax=axs[1][0])
sns.countplot(df['Category'],ax=axs[1][1])
axs[0][0].set_title("Ship Mode",fontsize=20)
axs[0][1].set_title("Segment",fontsize=20)
axs[1][0].set_title("Region",fontsize=20)
axs[1][1].set_title("Category",fontsize=20)
plt.tight_layout()
```

```python
state_profit=df.groupby(['Sub-Category'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Profit')
state_profit[['Profit']].plot(kind="bar",figsize=(10,10))
plt.ylabel("Profit")
plt.show()
```

```python
state_profit=df.groupby(['Sub-Category'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Discount')
state_profit[['Discount']].plot(kind="bar",figsize=(15,4))
plt.ylabel("Discount")
plt.show()
```

```python
state_profit=df.groupby(['Sub-Category'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Sales')
state_profit[['Sales']].plot(kind="bar",figsize=(15,4))
plt.ylabel("Sales")
plt.show()
```

```python
state_profit=df.groupby(['State'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Profit')
state_profit[['Profit']].plot(kind="bar",figsize=(15,4),color="purple")
plt.ylabel("Profit")
plt.show()
```

```python
state_profit=df.groupby(['State'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Discount')
state_profit[['Discount']].plot(kind="bar",figsize=(15,4),color="purple")
plt.ylabel("Discount")
plt.show()
```

```python
state_profit=df.groupby(['State'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Sales')
state_profit[['Sales']].plot(kind="bar",figsize=(15,4),color="purple")
plt.ylabel("Sales")
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
sns.distplot(df['Profit'],ax=axs[0][0],color="red")
sns.distplot(df['Discount'],ax=axs[0][1],color="blue")
sns.distplot(df['Sales'],ax=axs[1][0],color="yellow")
sns.distplot(df['Quantity'],ax=axs[1][1],color="green")
axs[0][0].set_title("Profit",fontsize=20)
axs[0][1].set_title("Discount",fontsize=20)
axs[1][0].set_title("Sales",fontsize=20)
axs[1][1].set_title("Quantity",fontsize=20)
plt.tight_layout()
```
